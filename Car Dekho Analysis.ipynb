```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cars.csv')
df.head()
```

1) Can you find that there has been an improvement in mpg over the years ?


```python
x=df['mpg']
y=df['year']
plt.bar(y,x,color='g')
plt.xlabel('Y e a r')
plt.ylabel('m p g')
plt.title('Trends in mpg over the years')
```

As seen from the above bar graph the mpg has increased slightly from the year 1970 to 1984.

2) Does cubic inches of a vehicle affect the vehicle's time to reach 60 Miles per hour?

```python
x=df['cubicinches']
y=df['time-to-60']
plt.scatter(x,y)
plt.xlabel('Cubicinches')
plt.ylabel('Time-to-60 (secs)')
plt.title('Cubic inches vs Time-to-60')
```

From the above graph as the Cubic Inches is increasing the time to reach 60 km/h is decreasing.

3) Does a specific country prefer vehicles with more horsepower ?


```python
def avg(q):
    df1=df[df['brand']== q]
    sum1=0
    for x in range(0,len(df1)):
        sum1= sum1 + df1.hp.iloc[x]
    avg1=sum1/len(df1)
    return avg1

plt.barh(df.brand.unique(),[avg('US.'),avg('Europe.'),avg('Japan.')],color=['r','g','orange'])
plt.ylabel('Countries')
plt.xlabel('Horsepower')
plt.title('Avg horsepower preferred by countries')
plt.grid(linestyle="--",linewidth=.5)
```

From the above data, it is clear that U.S prefers more horsepower than other countries.

4) How does horsepower affect the average of the vehicle ?

```python
x=df['hp']
y=df['mpg']
plt.bar(x,y)
plt.xlabel('Horsepower')
plt.ylabel('Average')
plt.title('average vs horsepower')
```

From the above the graph it is clear that average decreases as the horsepower increases.

5) How does the cubic inches affect the average of the vehicle ?

```python
x=df['cubicinches']
y=df['mpg']
plt.scatter(x,y,marker='.',c='g')
plt.xlabel('Cubic inches')
plt.ylabel('Average')
plt.title('Cubic inches vs Average')
```

From above data it is clear as the average decreases with increase in horsepower.

6) What is the effect of the number of cylinders on the mpg ?


```python
sns.boxplot(x=df.cylinders,y=df.mpg,data=df)
plt.title("Average vs Cylinders")
plt.show()
```

From above data it can be concluded that the vehicles with 4 cylinders have better average than the vehicles having more than 4 cylinders.

7) What relation between cubic inches , number of cylinders and horsepower can you observe from the data ?

```python
plt.subplot(2,2,1)
sns.boxplot(x=df.cylinders,y=df.cubicinches,data=df)

plt.subplot(2,2,2)
sns.scatterplot(x=df.cubicinches,y=df.hp,data=df,hue='cubicinches')

plt.subplot(2,2,3)
sns.boxplot(x=df.cylinders,y=df.hp,data=df)

plt.subplots_adjust(left=-1, bottom=0,top=1.5, wspace=.3,hspace=0.4)
```

From the above graphs, it can be concluded that:

a. The Cubic inches increases as the number of cylinders increases.

b. The horsepower increases as the cubic inches increases.

c. The horsepower increases as the number of cylinders increases.

8) How much proportion of the total records is occupied by vehicles with 8 cylinders?


```python
x=df['cylinders']==8
plt.pie(x.value_counts(True),autopct="%.1f%%")
plt.legend(['non 8 cylinders','8 cylinders'],bbox_to_anchor=(1.25, 1))
```

From the above chart it can be concluded that 29.1% of vehicles have 8 cylinders.
